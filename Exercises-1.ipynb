### List the supporting roles (having n=2) played by Cary Grant in the 1940s, in order by year.

```python
#cast[(cast.name == 'Cary Grant') & (cast.n == 2) & ((cast.year > 1939) & (cast.year < 1950))].sort('year')
cast[(cast.name == 'Cary Grant') & (cast.n ==2) & (cast.year//10 == 194)].sort('year')
```

### List the leading roles that Cary Grant played in the 1940s in order by year.

```python
#cast[(cast.name == 'Cary Grant') & (cast.n == 1) & ((cast.year > 1939) & (cast.year < 1950))].sort('year')
cast[(cast.name == 'Cary Grant') & (cast.n ==1) & (cast.year // 10 == 194)].sort('year')
```
